# Generic Fixed Point Iteration

```python
def fp_iteration(f, x_0, N):
    """diese Funktion nimmt eine mathematische Funktion f
    und gibt eine Liste mit den Werten der Fixpunktiteration zurück
    mit dem Parameter N wird die Anzahl der Iterationen festegelegt
    """
    n = 0
    l = [x_0]  # Liste mit allen Iterationen
    while n < N:
        x_new = f(l[-1])  # neue Iterierte wird berechnet
        l.append(x_new)
        n += 1
    return l
```

```python
def fp_recursion(f, l, N, n=0):
    if n >= N:
        return l
    l.append(f(l[-1]))
    print(l)
    n += 1
    return fp_recursion(f, l, N, n)
```

# Cobwebbing
Wir beginnen mit ein wenig Theorie und was man beachten muss (kann), wenn man Cobwebbing selbst programmieren möchte. 
Beispiele wie das am Ende mit verschiedenen Funktionen ausschauen kann gibt es ganz unten.

<h3>Vordefinierte Funktionen:</h3>

```python
def plot_function(
    function,
    show_axis=False,
    with_id=True,
    extend_id=True,
    equal_axis=True,
    show_fixed_points=False,
    fp_precision=6,
):
    x_val = np.linspace(I[0], I[1], z)
    y_val = function(x_val)
    np.array([x_val, y_val])
    id_val = x_val
    if with_id:
        if extend_id:
            id_max = np.max([x_val, y_val])
            id_min = np.min([x_val, y_val])
            id_val = np.linspace(id_min, id_max, z)
        plt.plot(id_val, id_val, color="gray", label="x=y")
    if show_axis:
        plt.plot(np.zeros(z), id_val, color="gray", lw=0.5)
        plt.plot(id_val, np.zeros(z), color="gray", lw=0.5)
    if equal_axis:
        plt.axis("equal")
    plt.plot(x_val, y_val, color="purple", label=f"{function.__name__}(x)")
    plt.xlabel("x")
    plt.ylabel(f"{function.__name__}(x)")
    plt.legend()
    if show_fixed_points:
        fp_array = x_val[fixpoints(x_val, y_val, fp_precision)]
        plt.plot(fp_array, fp_array, "ko", label="Fixpunkt(e)")


def fixpoints(x_val, y_val, precision):
    diff = x_val - y_val
    asign = np.sign(np.round(diff, precision))
    l = (asign[1:] - asign[:-1] != 0).astype(int)
    return np.argwhere(l)


def cobwebb(function, steps, x0, halt_if_outside_Interval=True, print_list=True):
    x_curr = x0
    L = []
    for i in range(steps + 1):
        if halt_if_outside_Interval:
            if I[0] > x_curr or I[1] < x_curr:
                print(
                    "Werte außerhalb des Definitionsbereiches, Berechnung abgebrochen"
                )
                break
        L.append(x_curr)
        if print_list:
            print(f"{i}-te Iteration: \tx_{i}={x_curr}")
        x_curr = function(x_curr)  # iteration step
    L.append(x_curr)
    plot_cobwebbing(L)
    return


def plot_cobwebbing(Folge_xk, AusgabePlot=plt, num=10):
    y_prev = 0
    if np.size(Folge_xk) < 2:
        print(f"Folge {Folge_xk} zu kurz.")
        return
    AusgabePlot.plot(
        np.full(num, Folge_xk[0]),
        np.linspace(y_prev, Folge_xk[1], num),
        color="black",
        linestyle="--",
        label="cobwebbing",
    )
    AusgabePlot.legend()
    for i in range(np.size(Folge_xk) - 1):
        x_i = Folge_xk[i]
        x_iplus1 = Folge_xk[i + 1]
        AusgabePlot.plot(
            np.full(num, x_i),
            np.linspace(y_prev, x_iplus1, num),
            color="black",
            linestyle="--",
        )
        y_prev = x_iplus1
        AusgabePlot.plot(
            np.linspace(x_i, x_iplus1, num),
            np.full(num, y_prev),
            color="black",
            linestyle="--",
        )
    return
```

| optionale Variablen       | Werte (von,bis)  | Beschreibung                               |
|----------|------------|---------------|
|show_axis        |  True, False     | Zeigt Achsen im Plot an                    |
|with_id          |  True, False     | Zeigt x=y Linie im Plot                    |
|extend_id        |  True, False     | Verlängert x=y Linie (kann schöner aussehen)|
|equal_axis       |  True, False     | Skaliert x und y Achse gleich|
|show_fixed_points| True,False       | Zeigt Fixpunkte mit schwarzem Punkt im Plot|
|fp_precision     | int, 0-15        |Genauigkeit (Rundung) bei Fixpunktsuche|
halt_if_outside_Interval|True, False| Hält Cobwebbing automatisch an, falls das nächste x_k den Definitionsbereich verlassen würde
print_list|True, False| Druckt die Liste der Iterationsschritte aus

<h4>Beispiel: Fixpunkterkennung</h4>
Es kann passieren, falls die Funktion sich dem Fixpunkt sehr flach annähert, dass entweder KEIN Fixpunkt, oder mehrere an der selben stelle angezeigt werden. Um diesen Umstand zu beheben, kann die Variable fp_precision (Standardwert=6) verändert werden. Das Prinzip der Fixpunkerkennung ist wie im Theorem über die Existenz von NST. Diese Variable legt dabei fest, auf wieviele Stellen gerundet wird. (werden zu viele "Fixpunkte" erkannt kann diese zahl erhöht werden, wenn ein Fixpunkt nicht erkannt wird kann probiert werden, den wert zu verkleinern.)

```python
I = [0.95, 1.05]
plot_function(
    lambda x: x**3 - 2 * x**2 + 2 * x, show_fixed_points=True
)  # fp_percision ist standardweise auf 6 stellen
```

```python
plot_function(
    lambda x: x**3 - 2 * x**2 + 2 * x, show_fixed_points=True, fp_precision=7
)  # bei 7 stellen wird der Fixpunkt nicht mehr erkannt
```

```python
plot_function(
    lambda x: x**3 - 2 * x**2 + 2 * x, show_fixed_points=True, fp_precision=5
)  # bei 5 stellen ist die ungenauigkeit größer
```
